```python
feed_shapes = gt.shapes.build_geometry_by_shape(abq_feed)
```

```python
feed_shapes
```

```python
ids = []
lines = []

for key, coords in feed_shapes.items():
    lines.append(coords)
    ids.append(key)
```

```python
route_geo_frame = geopandas.GeoDataFrame({'id': ids, 'geometry': lines})
```
